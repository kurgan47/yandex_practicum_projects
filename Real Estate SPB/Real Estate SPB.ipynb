## Предобработка данных

### Заполнение пропусков

Начнем с количества балконов. Если наличие балконов не указано, то их скорее всего нет, иначе человек бы указал эту информацию. Заполним пропуски нулями и поменяем тип данных на `int64`.

```python
data['balcony'] = data['balcony'].fillna(0).astype('int64')
data['balcony'].value_counts()
```

То же самое можно сказать про наличие парков и прудов поблизости, если бы они были, то скорее всего человек бы их указал. Заменим пропуски нулями, а тип данных заменим на `int64`.

```python
data[['parks_around_3000','ponds_around_3000']]=data[['parks_around_3000','ponds_around_3000']].fillna(0).astype('int64')
data[['parks_around_3000','ponds_around_3000']].info()
```

Посмотрим на колонку `is_apartment`

```python
data['is_apartment'].value_counts()
```

В ней мало заполненных значений, а апартаментов еще меньше, всего 50 штук. Скорее всего, если бы это были апартаменты, то человек бы это указал. Заменим пропуски на `False`.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
data['is_apartment'].value_counts()
```

Воспользуемся командой `to_datetime` чтобы перевести данные из строкового формата в формат дата/время. Поскольку часы, минуты и секунды заполнены нулями, ограничимся форматом **%Y-%m-%d** - год, месяц, день.

### Обработка названий населенных пунктов.

Посмотрим сколько названий населенных пунктов у нас в данных.

```python
len(data['locality_name'].unique())
```

Всего 365 населенных пунктов. Но в них есть неявные дубликаты. Все повторения встречаются в типах населенных пунктов, но не в названиях. Сначала разделим названия на тип населенного пункта и название. 

Посмотрим на наличие пустых значений.

```python
data['locality_name'].isna().sum()
```

Их всего 49 из 23699. Вряд ли кто-то купит недвижимость не зная, где она находится. Поэтому отбросим эти пустые значения.

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
data.info()
```

Создадим функцию, разделяющую колонку `locality_name` на тип населенного пункта и название.

```python
def place_type_name(locality_name):
    place_type = ''
    place_name = ''
    place_list = locality_name.split(' ') # разбиваем на слова по пробелу
    for i in range(len(place_list)):
        if place_list[i][0].isupper():
            place_type = ' '.join(place_list[:i])
            place_name = ' '.join(place_list[i:])
            break
    return place_type, place_name
    
```

Проверим, что функция работает

```python
for place in data['locality_name'].unique():
    print(place_type_name(place))
    break
    
```

Создадим колонки с типом населенного пункта и его названием.

```python
data['locality_type'],data['locality_name'] = zip(*data['locality_name'].map(place_type_name)) 
data[['locality_name','locality_type']].head()
```

Посмотрим теперь сколько названий населенных пунктов в `place_name`

```python
len(data['locality_name'].unique())
```

Их 305 вместо 365, получается 60 названий продублировались.

Посмотрим на типы населенных пунктов:

```python
data['locality_type'].value_counts()
```

Самая распространенноая ошибка это замена буквы `ё` на букву `е`. Исправим это.

```python
def e_to_jo_replace(place_type):
    if 'поселок' in place_type:
        place_type = place_type.replace('поселок', 'посёлок')
    return place_type
```

```python
data['locality_type'] = data['locality_type'].apply(e_to_jo_replace)
data['locality_type'].value_counts()
```

Стало лучше, теперь заменим <br>
`городской поселок` на `поселок городского типа`, 
`посёлок городского типа имени` на `посёлок городского типа`, 
`посёлок станции` на `посёлок`,
`посёлок при железнодорожной станции` на `посёлок`, 
`коттеджный посёлок` на `посёлок`, 
`садоводческое некоммерческое товарищество` на `садовое товарищество`

```python
place_type_dict = {
    'городской посёлок':'посёлок городского типа', 
    'посёлок городского типа имени':'посёлок городского типа', 
    'посёлок станции':'посёлок', 
    'посёлок при железнодорожной станции':'посёлок', 
    'коттеджный посёлок':'посёлок',
    'садоводческое некоммерческое товарищество':'садовое товарищество'
}

for place_type in place_type_dict:
    data['locality_type'] = data['locality_type'].str.replace(place_type,place_type_dict[place_type])

data['locality_type'].value_counts()
    
```

Теперь проверим дубликаты.

```python
data.duplicated().sum()
```

Ну вот, не нашлось ни одного дубликата.

* Пропуске в колонке апартаменты были заменены на *False*. Если бы это были аппартаменты, то скорее всего человек бы это указал.

### Категоризация по этажам

Создадим функцию, проверяющую относится ли этаж к первому, последнему или он в серединке.

```python
def floor_category(row):
    if (np.isnan(row['floor']) or np.isnan(row['floors_total'])) and row['floor'] != 1:
        return np.NaN
    elif row['floor'] == 1:
        return 'first'
    elif row['floor'] == row['floors_total']:
        return 'last'
    else:
        return 'other'
```

Создадим новую колонку с категорией этажа: первый, последний или средний и применим к нему функцию.

```python
data['floor_category'] = data.apply(floor_category, axis=1)
```

* Судя по медианам, в центре в 2016 году недвижимость в центре подорожала, а потом дешевела вплоть до 2019 года включительно.

Сравнение общих показателей и показателей в Петербурге.
 
 * Во многом общие показатели определяются показателями в Петербурге, потому что в Петербурге больше всего объявлений, корректнее бы было исключить Петербург или центр Петербурга из общих показателей.
 * В общем и в Петербурге чем больше комнат, тем дороже квартира. То же самое справедливо и для площади. Это логично, чем больше комнат, тем больше площадь. При этом все большие квартиры на обоих графиках это почти полностью квартиры из центра Петербурга.
 * Если посмотреть по медианам, то в центре Петербурга квартиры на последних этажах стоят так же как и на средних, а в общем, квартиры на средних этажах стоят дороже. Скорее всего это связано с малоэтажной застройкой в центре Петербурга, и там иметь квартиру под крышей престижней, лучше вид, возможно там есть пентхаусы на последних этажах, с панорамными окнами, стеклянной крышей и т.д. В общем дома выше, чем в центре Питера, там вид на средних этажах не сильно отличается от вида на последних, зато на последних этажах под чердаком может течь или промерзать потолок. Судя по средним и там и там квартиры на последних этажах дороже, скорее всего и туда и туда попали дорогие квартиры из центра Питера и завысили средние значения.
 * Если посмотреть на средние значения по дням недели, то и в Питере и в общем по пятницам публикуют объявление о дорогих квартирах. Скорее всего объявления о продаже дорогих квартир в центре были сделаны в пятницу, они и задрали среденее. По медианам такой зависимости нет.
 * В общем в 2015-2016годах стоимость квартир падает. А вот в центре Петербурга растет. Это скорее всего связано с кризисом 2014 года, когда после присоединения Крыма к России были введены санкции и нефть упала в цене. На очень богатых людях это видимо не так сильно отразилось, а может быть они решили сохранить свои деньги через покупку недвижимости, поэтому цена в центре Петербурга выросла. В 2018 и 2019 годах цена на нефть подросла и экономика начала восстанавливаться, поэтому и спрос на обычное жилье стал расти, но на покупку дорогого жилья в центре еще не хватает.

## Вывод

* Наиболее явная корреляция стоимости квартиры с площадью.
* Почти все нестандартные дорогие квартиры с большой площадью и большим количеством комнат находятся в центре Сенкт-Петербурга. Эти данные реальные, поэтому мы их не стали исключать.
* Чем дальше от центра, тем дешевле.
* Чем больше комнат, тем дороже стоит квартира. Это справедливо для квартир до 7 комнат.
* Квартиры дешевели с 2014 до 2018 года, а потом стали дорожать, в центре Питера наблюдается обратная картина, в 2016 году они подорожали, а потом стали дешеветь.
* Самые дешёвые квартиры на первом этаже, на средних этажах самые дорогие. В центре стоимость на последних и средних этажах не отличается и на последних даже может быть дороже.
