### Проверка правильности расчёта эффективности обогащения.

Эффективность обогащения рассчитывается по формуле:

$RECOVERY = \frac{C\cdot (F-T)}{F\cdot(C-T)}\times100\%$, где 
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.


```python
c = gold_train['final.output.concentrate_au']
f = gold_train['rougher.input.feed_au']
t = gold_train['final.output.tail_au']

recovery_final = c*(f-t)/(f*(c-t))*100
```

```python
plt.scatter(recovery_final,gold_train['final.output.recovery'],s=50)
plt.plot([0,100],[0,100],'-r',linewidth=4)
plt.title('RECOVERY calculated vs original')
plt.xlabel('calculated')
plt.ylabel('original')
plt.show()
```

Посчитаем МАЕ по формуле:

$MAE = \frac{\sum_{i=1}^{n}\lvert x_{i}-y_{i}\rvert}{n}$

```python
np.abs(recovery_final-gold_train['final.output.recovery']).sum()/len(recovery_final)
```

```python
mean_absolute_error(recovery_final,gold_train['final.output.recovery'])
```

Значит эффективность всего процесса посчитана правильно.

Посчитаем эффективность для фолтации

```python
c = gold_train['rougher.output.concentrate_au']
f = gold_train['rougher.input.feed_au']
t = gold_train['rougher.output.tail_au']

recovery_rougher = c*(f-t)/(f*(c-t))*100
```

```python
plt.scatter(recovery_rougher,gold_train['rougher.output.recovery'],s=50)
plt.plot([0,100],[0,100],'-r',linewidth=4)
plt.title('RECOVERY calculated vs original')
plt.xlabel('calculated')
plt.ylabel('original')
plt.show()
```

```python
np.abs(recovery_rougher-gold_train['rougher.output.recovery']).sum()/len(recovery_rougher)
```

```python
mean_absolute_error(recovery_rougher,gold_train['rougher.output.recovery'])
```

### Концентрация металлов (Au, Ag, Pb) на различных этапах очистки

Напишем функцию, которая будет строить гистограммы концентрации металлов на различных этапах

```python
def plot_metal(metal,c):
    fig, axes = plt.subplots(2, 2, figsize=(15,10))
    fig.suptitle(f'{metal} concentration at different phases')
    sns.histplot(data=gold_full, x=f'rougher.input.feed_{metal.lower()}',color=c,ax=axes[0,0])
    axes[0,0].set_title(f'Rougher input {metal}')
    sns.histplot(data=gold_full, x=f'rougher.output.concentrate_{metal.lower()}',color=c,ax=axes[0,1])
    axes[0,1].set_title(f'Rougher output {metal}')
    sns.histplot(data=gold_full, x=f'primary_cleaner.output.concentrate_{metal.lower()}',color=c,ax=axes[1,0])
    axes[1,0].set_title(f'Primary cleaner output {metal}')
    sns.histplot(data=gold_full, x=f'final.output.concentrate_{metal.lower()}' ,color=c,ax=axes[1,1])
    axes[1,1].set_title(f'Final output {metal}')
    plt.subplots_adjust(hspace=0.25)
    plt.show()
```

## Модель.

### sMAPE

Напишем функцию для рассчета sMAPE.

```python
def smape(y_true, y_pred):
    return 100*(1/len(y_true))*np.sum((np.abs(y_true-y_pred)/(np.abs(y_true)+np.abs(y_pred))/2))
```

```python
smape = make_scorer(smape, greater_is_better=False)
```
