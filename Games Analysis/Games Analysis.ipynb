# Описание проекта

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
В наличие есть данные до 2016 года. Необходимо дать прогноз на 2017 год.

## Загрузка и обработка данных

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid", color_codes='pastel')
from scipy import stats as st
from numpy.random import choice
```

```python
data = pd.read_csv('https://code.s3.yandex.net/datasets/games.csv')
data.head()
```

В названиях колонок присутсвуют заглавные буквы. Поменяем названия так, чтобы все буквы были маленькими.

```python
data.columns = data.columns.str.lower()
data.columns
```

Посмотрим информацию о таблице, командой info

```python
data.info()
```

Год выпуска игры типа float из-за того, что там присутствуют пустые значения. Поскольку наша задача подготовить прогноз на следующий год, и колонка с годом является важной, а пропусков там немного, то отбросим строчки с пустыми значениями и приведем колонку к типу int.

```python
data = data.dropna(subset=['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype('int32')
```

Колонка с рейтенгом пользователей должна принимать числовые значения, а она является строкой. Посмотрим почему.

```python
data['user_score'].value_counts()
```

```python
data.query('user_score=="tbd"').head()
```

Значение *tbd* встречается 2376 раз. Это расшифровывается как *to be determined*, т.е. он еще не определен. Поэтому заменим *tbd* на пустые значения. Видимо сначала рейтинг определют критики, а пользователи покупают игры и выставляют свой рейтинг позже.

```python
data['user_score'] = (data['user_score'].mask(data['user_score']=='tbd')).astype('float64')
data.info()
```

Есть пропуски в колонках с рейтингами. Лучше оставим их пустыми, потому что они могут принимать любые значения в указанных перделах и заполнять их средним или медианой будет некорректно.

Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.

```python
data['total_sales'] = data[['na_sales', 'eu_sales','jp_sales', 'other_sales']].sum(axis=1)
data = data[['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales',
'jp_sales', 'other_sales', 'total_sales','critic_score', 'user_score', 'rating']]
data.head()
```

```python
data.info()
```

```python
data = data.dropna(subset=['name','genre']).reset_index(drop=True)
```

Пропуски в колонке `rating` заполним значением 'unknown'.

```python
data['rating'].fillna('unknown',inplace=True)
```

```python
data['rating'].isna().sum()
```

```python
data.info()
```

### Выпуск игр по годам

Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

- Максимум произвоства игр приходится на 2008 - 2009 года. Дальше идет спад. Возможно это связано с развитием смартфонов и мобильного интернета, часть пользователей могли перейти на игры на гаджетах.

Хотя платформы появляются и исчезают примерно за 10 лет, не стоит брать 10 лет как актуальный срок для прогнозирования. В 2006 - 2007 годах были совсем другие приставки и совсем другие игры.

Посмотрим на продажи игр по платформам по годам. Даже если игры с этой платформы продаются в 2014-2015 годах, но в 2017 продаваться не будут, то такую платформу стоит исключить, ведь прогноз по ней 0.

### Влияние отзывов на продажи

Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

Самая попкулярная платформа у нас PS2, построим графики для нее.

```python
data_to_plot = pd.concat([data.query('platform=="PS4"'), data.query('platform=="PS4"')], keys=['critic', 'user'])
data_to_plot.reset_index(inplace=True)
data_to_plot = data_to_plot.rename(columns={"level_0": "score_provided"}).drop(columns='level_1')
data_to_plot['score']  = data_to_plot['critic_score']
data_to_plot.loc[data_to_plot['score_provided']=='user','score'] = data_to_plot.loc[data_to_plot['score_provided']=='user','user_score']
```

```python
sns.relplot(data=data_to_plot,
              x='score',y='total_sales',col='score_provided',facet_kws=dict(sharex=False),
              linewidth=0,hue='year_of_release', palette="magma")
```

```python
sns.relplot(data=data_to_plot,
              x='score',y='total_sales',kind='line',col='score_provided',facet_kws=dict(sharex=False),errorbar='sd')
```

Посчитаем коэффициент корреляции.

```python
data.query('platform=="PS4"')['critic_score'].corr(data.query('platform=="PS4"')['total_sales'])
```

```python
data.query('platform=="PS4"')['user_score'].corr(data.query('platform=="PS4"')['total_sales'])
```

- Корреляция между рейтингом и продаджами очень мальенька, что говорит об отсутсвии линейной связи между этими величинами.
- Тем не менее, в области высоких оценок критиков наблюдается увеличение продаж.
- От оценок пользователей зависимость слабее, но тоже с ростом оценок наблюдается увеличение продаж.

Посмотрим, справедливо ли то же самое для остальных платформ.

```python
data_to_plot = pd.concat([data.query('platform in @ platforms_to_use and year_of_release>=2012'), data.query('platform in @ platforms_to_use and year_of_release>=2012')], keys=['critic', 'user'])
data_to_plot.reset_index(inplace=True)
data_to_plot = data_to_plot.rename(columns={"level_0": "score_provided"}).drop(columns='level_1')
data_to_plot['score']  = data_to_plot['critic_score']
data_to_plot.loc[data_to_plot['score_provided']=='user','score'] = data_to_plot.loc[data_to_plot['score_provided']=='user','user_score']
```

```python
sns.relplot(data=data_to_plot,
              x='score',y='total_sales',row='platform',col='score_provided',facet_kws=dict(sharex=False),
              linewidth=0,hue='year_of_release', palette="magma")
```

```python
sns.relplot(data=data_to_plot,
              x='score',y='total_sales',hue='platform',kind='line',row='platform',col='score_provided',
            facet_kws=dict(sharex=False),errorbar='sd')
```

- Для большикства платформ справедливо, что при повышении оценок от критиков продажи возрастают.
- Такой явной зависимости для оценок пользователей не наблюдается.
- Видимо пользователи более эмоционально ставят оценки, а кртитики следуют более формальным критериям.
- Причиной наблюдаемого различия может быть то, что сначала критики выставляют оценки, потом пользователи глядя на эти оценки покупают игры и выставляют оценки уже по своим впечатлениям. Не зря же мы обрабатывали значение *tbd*.
- На WiiU обратная картина, при высоких оценках пользователей наблюдается рост продаж. Наверное хорошая платформа.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.


Сформулируем нулевую и альтернативныую гипотезы

- $H_0$ Средние пользовательские рейтинги платформ XOne и PC одинаковые
- $H_1$ Средние пользовательские рейтинги платформ XOne и PC отличаются

Для начала посмотрим на распределения.

Создадим выборки для теста.

```python
sns.histplot(data=data.query('platform in ["PC","XOne"]'), x="user_score", hue="platform",hue_order=['XOne','PC'])
```

```python
platforms_by_year = pd.pivot_table(data,values='total_sales',index = 'year_of_release',columns='platform',aggfunc='sum')
platforms_by_year[['XOne','PC']].dropna(how='all').plot(kind='bar', stacked=True, width=1.0,figsize=(7,4))
plt.ylabel('total_sales')
plt.show()
```

```python
pc = data.query('platform=="PC" and year_of_release>=2013')['user_score'].dropna()
xone = data.query('platform=="XOne"')['user_score'].dropna()
```

```python
len(pc)
```

```python
len(xone)
```

```python
pc.mean()
```

```python
xone.mean()
```

Будем считать, что люди, которые играют на PC не играют в XOne и наоборот, тогда выборки будут независимыми.

```python
results = st.ttest_ind(pc, xone)
results.pvalue
```

Может быть одни и те же геймеры играют на разных платформах и оценивают и те и другие, тогда выборки будут зависимыми.

```python
results = st.ttest_rel(pc,xone.sample(n=155, random_state=1))
results.pvalue
```

Возьмем пороговое значение 0.05. Потому что это общепринятое пороговое значение. Вероятность получить наблюдаемые средние при условии, что верна нулевая гипотеза равна ~14%, она выше, чем пороговое значение. Принимаем нулевую гипотезу.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем нулевую и альтернативную гипотезы.
- $H_0$ Средние пользовательские рейтинги жанров Action и Sports равны.
- $H_1$ Средние пользовательские рейтинги жанров Action и Sports отличаются.

```python
sns.histplot(data=data.query('genre in ["Action","Sports"]'), x="user_score", hue="genre")
```

Создадим выборки для проверки гипотезы.

```python
action = data.query('genre=="Action"')['user_score'].dropna()
sports = data.query('genre=="Sports"')['user_score'].dropna()
```

```python
action.mean()
```

```python
sports.mean()
```

```python
len(action)
```

```python
len(sports)
```

Будем считать, что одни и те же пользователи играют в игры разных жанров, тогда выборки будут зависимыми.

```python
results = st.ttest_rel(action.sample(n=1081, random_state=1),sports)
results.pvalue
```

Вероятность получить наблюдаемую разницу между средними двух выборок по дпнным при условии, что нулевая гипотеза верна, равна ~11%, что достаточно много, и выше порогового значения в 5%. Принимаем нулевую гипотезу.

Возможно сначала критики выставляют свои рейтинги, потом пользователи покупают игры основываясь на рейтингах критиков, а затем выставляют свои рейтинги, основываясь на впечатлениях от игры.
- У американцев и европейцев предпочтения по платформам совпадают, самые популярные PS4 и XOne, на треьем месте 3DS. Японцы больше ценят мобильность и предпочитают портативные консоли, у них популярны 3DS и ЗPSV.
- Самые популярные жанры у американцев и европейцев экшн, шутеры, спорт и ролевые игры. У японцев на первом месте ролевые игры, экшн, а спорт вообще не входит в топ 5.
- Самая популярная возростная категория - для всех, в основном в игры играют подростки.
- Средний пользовательский рейтинг на PC ниже, чем на XBox. Возможно для игр нужны очень мощьные и дорогие компьютеры, а если компьютер не тянет, то и впечатления от игры хуже, а XBox изначально сделан для игр.
- Средние пользовательские рейтинги экшн и спортивных игр совпадают. В спортивные игры играет меньше пользователей, но видимо эмоций от спортивных игр они получают не меньше, чем от экшн игр. В спортивные игры как правило играют в компании, возможно накладывается еще и впечатление от общения, поэтому и средний рейтинг такой же высокий как у экшн игр.


- В Америке популярны категории M,E10+,E и T.
- В Европпе популярны категории М, E10+, E и T.
- В Японии система рейтингов работает не так хорошо, как в Аменрике и Европпе.
- Средние пользовательские рейтинги на PC и на XOne равны. Хорошая игра хороша на любой платформе, а плохая не станет хорошей от смены платформы.
