## Описание проекта

Из «Бета-Банка» стали уходить клиенты. Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Имеются исторические данные о поведении клиентов и расторжении договоров с банком. 

## Описание данных

**Признаки**
- RowNumber — индекс строки в данных
- CustomerId — уникальный идентификатор клиента
- Surname — фамилия
- CreditScore — кредитный рейтинг
- Geography — страна проживания
- Gender — пол
- Age — возраст
- Tenure — сколько лет человек является клиентом банка
- Balance — баланс на счёте
- NumOfProducts — количество продуктов банка, используемых клиентом
- HasCrCard — наличие кредитной карты
- IsActiveMember — активность клиента
- EstimatedSalary — предполагаемая зарплата

**Целевой признак**
- Exited — факт ухода клиента


## Импорт библиотек

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

import re
import seaborn as sns
sns.set_theme(style="ticks")
import matplotlib.pyplot as plt
```

## Загрузка и анализ данных

```python
data = pd.read_csv('https://code.s3.yandex.net/datasets/Churn.csv')
data.head()
```

Прииведем названия колонок к стандартному виду.

```python
col_dict = {}
for col in data.columns:
    col_dict[col]='_'.join(re.findall('[A-Z][a-z]*', col)).lower()
data = data.rename(columns=col_dict)
data.columns
```

Номер раяда нам не нужен,фамилия  и id клиента тоже вряд ли пригодятся, удалим эти данные.

```python
data = data.drop(columns=['row_number','surname','customer_id'])
```

```python
data.info()
```

 Заполним пропуски медианой.

```python
data['tenure'] = data.tenure.fillna(data.groupby(['gender','geography'])['tenure'].transform('median'))
data['tenure'].isna().sum()
```

## Подготовка данных

Подготовим категориальные признаки методом OHE.

```python
data_ohe = pd.get_dummies(data,drop_first=True)
data_ohe.columns = data_ohe.columns.str.lower()
data_ohe.columns
```

Разделим данные на признаки и целеывые признаки.

```python
features = data_ohe.drop(columns=['exited'])
target = data_ohe['exited']
```

Посмотрим на дисбаланс классов.

```python
target.value_counts()
```

Клиенты покидают банк примено в 4 раза реже, чем остаются.

Разделим данные на обучающую, валидационную и тестовые выборки в соотношении 3:1:1.

```python
features_train, features_check, target_train, target_check = train_test_split(features,target,test_size=0.4,random_state=12345,stratify=target)
features_valid, features_test, target_valid, target_test = train_test_split(features_check,target_check,
                                                                            test_size=0.5,random_state=12345,stratify=target_check)
print(features_train.shape, features_valid.shape, features_test.shape)
```

Выделим количественные признаки и произведем их стандартизацию.

```python
numeric = ['credit_score','balance','age','estimated_salary','tenure']
```

```python
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train_scaled = features_train.copy()
features_train_scaled[numeric] = scaler.transform(features_train_scaled[numeric])
features_train_scaled.head()
```

```python
features_valid_scaled = features_valid.copy()
features_valid_scaled[numeric] = scaler.transform(features_valid_scaled[numeric])
```

```python
features_test_scaled = features_test.copy()
features_test_scaled[numeric] = scaler.transform(features_test_scaled[numeric])
```

## Модель без учета дисбаланса

### Дерево решений

Посмотрим на метрики F1 и площадь под ROC кривой при различной глубине дерева решений.

```python
f1 = []
auc_roc = []
for i in range(2,30):
    tree = DecisionTreeClassifier(max_depth = i, random_state=12345)
    tree.fit(features_train,target_train)
    prediction_valid = tree.predict(features_valid)
    f1.append(f1_score(target_valid, prediction_valid))
    probabilities_one_valid =tree.predict_proba(features_valid)[:,1]
    auc_roc.append(roc_auc_score(target_valid, probabilities_one_valid))
    print(i,f1_score(target_valid, prediction_valid),roc_auc_score(target_valid, probabilities_one_valid))
```

```python
fig = plt.figure(figsize=(13, 6))
plt.subplot(1, 2, 1)

plt.plot(range(2,30),f1)
plt.title('Decision tree F1 score')
plt.ylabel('F1')
plt.xlabel('Tree depth')

plt.subplot(1, 2, 2)
plt.plot(range(2,30),auc_roc)
plt.title('Decision tree AUC_ROC score')
plt.ylabel('AUC_ROC')
plt.xlabel('Tree depth')
```

Наибольшее значение F1 0.58 на валидационной выборке достигается при глубине дерева 7, а наибольшее занчение площади под ROC кривой 0.84 при глубине дерева равной 5.

Посмотрим изменится ли тоность прогноза, если использовать масштабируемые данные.

```python
f1 = []
auc_roc = []
for i in range(2,30):
    tree = DecisionTreeClassifier(max_depth = i, random_state=12345)
    tree.fit(features_train_scaled,target_train)
    prediction_valid = tree.predict(features_valid_scaled)
    f1.append(f1_score(target_valid, prediction_valid))
    probabilities_one_valid = tree.predict_proba(features_valid_scaled)[:,1]
    auc_roc.append(roc_auc_score(target_valid, probabilities_one_valid))
    print(i,f1_score(target_valid, prediction_valid),roc_auc_score(target_valid, probabilities_one_valid))
```

```python
fig = plt.figure(figsize=(13, 6))
plt.subplot(1, 2, 1)

plt.plot(range(2,30),f1)
plt.title('Decision tree F1 score')
plt.ylabel('F1')
plt.xlabel('Tree depth')

plt.subplot(1, 2, 2)

plt.plot(range(2,30),auc_roc)
plt.title('Decision tree AUC_ROC score')
plt.ylabel('AUC_ROC')
plt.xlabel('Tree depth')
```

### Случайный лес

## Модель с учетом дисбаланса

### Дерево решений

#### Взвешивание классов

Обучим модель с весами сбалансированными по классам. Наибольшие значения метрик F1 и AUC_ROC были при глубине дерева равной 5.

```python

for i in range(2,30):
    tree = DecisionTreeClassifier(max_depth = i, random_state=12345, class_weight='balanced')
    tree.fit(features_train,target_train)
    prediction_valid = tree.predict(features_valid)
    probabilities_one_valid =tree.predict_proba(features_valid)[:,1]
    print(i,f1_score(target_valid, prediction_valid),roc_auc_score(target_valid, probabilities_one_valid))
```

До 0,59 не хватает.

#### Увеличение выбрки

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(
    features_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled
```

Попробуем увеличить количество класса 1 в обучающей выборке в 4 раза и подобрать гиперпараметры заново. 


```python
features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
for i in range(2,15):
    tree = DecisionTreeClassifier(max_depth = i, random_state=12345)
    tree.fit(features_upsampled,target_upsampled)
    prediction_valid = tree.predict(features_valid)
    probabilities_one_valid =tree.predict_proba(features_valid)[:,1]
    print(i,f1_score(target_valid, prediction_valid),roc_auc_score(target_valid, probabilities_one_valid))
```

#### Уменьшение выборки

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_downsampled = pd.concat(
    [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
    [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    features_downsampled, target_downsampled = shuffle(
    features_downsampled, target_downsampled, random_state=12345)
    return features_downsampled, target_downsampled
```

Попробуем уменьшить выборку в 4 раза

```python
features_downsampled,target_downsampled = downsample(features_train, target_train, 0.25)
```

```python
for i in range(2,15):
    tree = DecisionTreeClassifier(max_depth = i, random_state=12345)
    tree.fit(features_downsampled,target_downsampled)
    prediction_valid = tree.predict(features_valid)
    probabilities_one_valid =tree.predict_proba(features_valid)[:,1]
    print(i,f1_score(target_valid, prediction_valid),roc_auc_score(target_valid, probabilities_one_valid))
```

При глубине дерева 6 значение F1 равно 0.59.

#### Изменение порога

```python
tree = DecisionTreeClassifier(max_depth = 5, random_state=12345)
tree.fit(features_train,target_train)
probabilities_one_valid =tree.predict_proba(features_valid)[:,1]
f1=[]
for value in np.arange(0, 1, 0.05):
    prediction_valid = probabilities_one_valid>=value
    print(value,f1_score(target_valid, prediction_valid))
    f1.append(f1_score(target_valid, prediction_valid))
```

При отсечке 0.3 и 0.35 поличили наилучшее значение F1 равное 0.6

```python
plt.plot(np.arange(0, 1, 0.05),f1)
plt.title('Decision tree F1')
plt.ylabel('F1')
plt.xlabel('threshold')
```

При отсечке 0.25 достигается значение метрики F1 0.61 на валидационной выборке выше порогового значения 0.59.

Посмотрим, где эта точка находится на ROC кривой.
