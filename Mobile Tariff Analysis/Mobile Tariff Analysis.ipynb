# Описание проекта

Есть два тарифных плана: «Смарт» и «Ультра». Нужно понять, какой тариф приносит больше денег.

# Описание тарифов

**Тариф «Смарт»**
1.	Ежемесячная плата: 550 рублей
2.	Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
3.	Стоимость услуг сверх тарифного пакета: 
- минута разговора: 3 рубля
- сообщение: 3 рубля
- 1 Гб интернет-трафика: 200 рублей

**Тариф «Ультра»**
1.	Ежемесячная плата: 1950 рублей
2.	Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
3.	Стоимость услуг сверх тарифного пакета: 
- минута разговора: 1 рубль
- сообщение: 1 рубль
- 1 Гб интернет-трафика: 150 рублей

**Примечание**
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта. Примите, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

## Подготовка данных

### Переименование столбцов

В таблицах переименуем колонку `id` в соответсвии с названием таблицы. Это позволит избежать путаницы.

```python
calls = calls.rename(columns = {'id':'call_id'})
calls.columns
```

```python
internet = internet.rename(columns = {'id':'internet_id'})
internet.columns
```

```python
messages = messages.rename(columns = {'id':'message_id'})
messages.columns
```

### Округление длительности звонков до целых значений

По правилам Мегалайн длительность звонков округляется до минут в большую сторону.

```python
display(calls.head())
calls['duration'] = (calls['duration'].apply(np.ceil)).astype('int64')
calls.head()
```

### Приведение `id` к типу *int*

Во всех колонках `id` строковый тип данных формата `user_id_число`, а `user_id` хранится в отдельной колонке. Можно без потерь привести данные из строки в целое число.  

```python
calls['call_id'] = calls['call_id'].str.split(pat='_',expand=True)[1].astype('int64')
calls.head()
```

```python
internet['internet_id'] = internet['internet_id'].str.split(pat='_',expand=True)[1].astype('int64')
internet.head()
```

```python
messages['message_id'] = messages['message_id'].str.split(pat='_',expand=True)[1].astype('int64')
messages.head()
```

### Количество звонков в месяц и их длительность

Посчитаем для каждого пользователя количество сделланых звонков по месяцам.

```python
calls['month'] = pd.DatetimeIndex(calls['call_date']).month
```

```python
calls_pivot = pd.pivot_table(calls,values='duration',index=['user_id','month'],aggfunc=['count','sum'],fill_value=0)
calls_pivot.head()
```

```python
calls_pivot.columns = ['calls_number','calls_duration']
calls_pivot.head()
```

### Количество СМС в месяц

```python
messages['month'] = pd.DatetimeIndex(messages['message_date']).month
```

```python
messages_pivot = pd.pivot_table(messages,values='message_date',index=['user_id','month'],aggfunc='count',fill_value=0)
messages_pivot.head()
```

```python
messages_pivot.columns = ['sms']
messages_pivot.head()
```

### Объем израсходованного интернет-трафика по месяцам

```python
internet['month'] = pd.DatetimeIndex(internet['session_date']).month
```

```python
internet_pivot = pd.pivot_table(internet,values='mb_used',index=['user_id','month'],aggfunc='sum',fill_value=0)
internet_pivot.head()
```

Поскольку Мегалайн жадный оператор, переведем мегабайты в гигабайты с округлением в большую сторону.

```python
internet_pivot.columns = ['gb_used']
internet_pivot['gb_used'] = ((internet_pivot['gb_used']/1024).apply(np.ceil)).astype('int64')
```

```python
internet_pivot.head()
```

Ксения Агаева купила тариф "Ультра" 15 декабря, а 23 декабря отказалась от его использования. Можно сказать, что никаких данных мы не потеряли.

Напишем функцию, которая считает траты за месяц.

```python
def monthly_cost(row):
    if row['tariff'] == 'smart':
        cost = 550
        gb = 15
        mins = 500
        sms = 50
        extra_gb_cost = 200
        extra_sms_cost = 3
        extra_min_cost = 3
    elif row['tariff'] == 'ultra':
        cost = 1950
        gb = 30
        mins = 3000
        sms = 1000
        extra_gb_cost = 150
        extra_sms_cost = 1
        extra_min_cost = 1
    monthly_fee = cost+max(0,row['calls_duration']-mins)*extra_min_cost+max(0,row['sms']-sms)*extra_sms_cost+max(0,row['gb_used']-gb)*extra_gb_cost
    return monthly_fee
```

Применим ее к нашим данным.

```python
data['monthly_fee'] = data.apply(monthly_cost,axis=1)
data.head()
```

## Проверка гипотез

### Гипотеза: средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

Построим гистограммы по тарифам и посмотрим на распределение выручки.

```python
ax=sns.displot(data, x='monthly_fee',hue='tariff',palette='pastel', 
               edgecolor='.6',log_scale=(False,True))
```

Наблюдаются пики на уровне ежемесячной абонентской платы. Тем не менее многие пользователи не укладываются в абонетскую плату и платят больше.

Посмотрим на общую и среднюю выручку по тарифам

```python
pd.pivot_table(data,index='tariff',values='monthly_fee',aggfunc=['sum','mean','var','std','count'])
```

Общаяя выручка по тарифу "Смарт" больше, чем по тарифу "Ультра". Но количество пользователей тарифа "Смарт" превышает количество пользователей "Ультра", и значение среденей вырочки по Смарту меньше.

Сформулируем нулевую $H_0$ и альтернатиыную $H_1$ гипотезы:
- $H_0$ Средняя выручка пользователей по тарифу "Смарт" равна средней выручке пользователей по тарифу "Ультра".
- $H_1$ Средняя выручка пользователей по тарифу "Смарт" меньше средней выручки пользователей по тарифу "Ультра".

Посмотрим на результаты теста, если у тарифов разные распределения. Будем считать, что распределения не зависят друг от друга, потому что это разные тарифы и пользуются ими разные люди.

```python
smart_fee = data.query('tariff=="smart"')['monthly_fee']
ultra_fee = data.query('tariff=="ultra"')['monthly_fee']
results = st.ttest_ind(smart_fee,ultra_fee, equal_var = False)
results.pvalue
```

А если взять одинаковые распределения?

```python
results = st.ttest_ind(smart_fee,ultra_fee)
results.pvalue
```

А если выборку одинаковой длины?

```python
results = st.ttest_ind(smart_fee.sample(len(ultra_fee)),ultra_fee)
results.pvalue
```

#### Вывод

 - Не важно какое пороговое значение мы возьмем 5% или 1%, одинаковой или разной длины выборки, одинаковые или разные распределения для выборок, всё равно уровень значимости ниже порогового значения.
 - Отвергаем нулевую гипотезу о том, что среднее значение выручки одинаковое.
 - Принимаем альтернативную гепотезу о том, что среднее значение вырочки по тарифу "Смарт" меньше, чем среднее значение выручки по тарифу "Ультра".
 

### Гипотеза: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Разделим пользователей по городам. Выделим две группы, пользователей из Москвы и из других городов.

```python
def is_moscow(city):
    if city == 'Москва':
        return 'Moscow'
    else:
        return 'Other'
```

```python
data['moscow'] = data['city'].apply(is_moscow)
data['moscow'].unique()
```

```python
ax=sns.displot(data, x='monthly_fee',hue='moscow',hue_order=['Moscow','Other'],log_scale=(False,True),palette='pastel', 
               edgecolor='.6')
```

Распределения в Москве и в других городах похожи друг на друга и отличаются в основном количесвом пользователей.

```python
pd.pivot_table(data, index='moscow', values='monthly_fee',aggfunc=['sum','mean','var','std','count'])
```

Сформулируем нулевую $H_0$ и альтернативную $H_1$ гипотезы.
- $H_0$ Средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов.
- $H_1$ Средняя выручка пользователей из Москвы отличается от средней выручки пользователей из других регионов.

Мы оцениваем одни и те же тарифы, но на разной территории. Будем считать, что выборки не независимые.

Разделим датасет на Москву и не Москву. 

```python
moscow = data.query('moscow=="Moscow"')
other = data.query('moscow=="Other"')
```

Статистический тест для зависимых выборок принимает на вход только выборки одинаковой длины. Пользователей в Москве меньше, чем в основных регионах. Выборкку по Москве возьмем полностью, а по остальным регионам возьмем случайную выборку такой же длины как московская выборка.

Чтобы количество образцов в выборке было пропроционально количеству пользоваелей в выборке используем параметр weights в функции sample. Проверяющий работы про кредиты Сороколетов Илья рекомендовал в таких случаях использовать функцию transform.

```python
other_fee_sample = other.sample(len(moscow),weights = other.groupby('city')['city'].transform('count'))['monthly_fee']
moscow_fee_sample = moscow['monthly_fee']
```

```python
results = st.ttest_rel(moscow_fee_sample,other_fee_sample)
results.pvalue
```

А если взять независимые выборки?

```python
results = st.ttest_ind(moscow_fee_sample,other_fee_sample)
results.pvalue
```

А независимые выборки изначального размера?

```python
results = st.ttest_ind(moscow['monthly_fee'],other['monthly_fee'])
results.pvalue
```

#### Вывод
- Не важно какое пороговое значение мы возьмем 1% или 5% и будем ли считать выборки зависимыми или независимыми, все равно уровень значимости выше прогового значения.
- Принимаем нулевую гипотезу о том, что средняя выручка в Москве равна средней выручке в других регионах.
